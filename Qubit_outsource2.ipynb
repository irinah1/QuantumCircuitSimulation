```python
q2 = Qubit([1, 0, 0])
print("Qubit:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(amplitude_damping(q2.get_Bloch_vector(), 0.005, 9))
print("Amplitude damping:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(phase_damping(q2.get_Bloch_vector(), 0.008, 9))
print("Phase damping:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(bit_flip(q2.get_Bloch_vector(), 0.05))
print("Bit flip:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(phase_flip(q2.get_Bloch_vector(), 0.05))
print("Phase flip:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(bit_phase_flip(q2.get_Bloch_vector(), 0.05))
print("Bit-phase flip:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())

q2._set_Bloch_vector(depolarization(q2.get_Bloch_vector(), 0.05))
print("Depolarization:")
print(q2.get_matrix())
print(q2.get_Bloch_vector())
```

```python
# quibit history
q2 = Qubit([1, 0, 0])
number_time_steps = 10
xarr = []
yarr = []
zarr = []
for t in range(number_time_steps):
    q2._set_Bloch_vector(amplitude_damping(q2.get_Bloch_vector(), 0.005, t))
    q2._set_Bloch_vector(phase_damping(q2.get_Bloch_vector(), 0.008, t))
    x, y, z = q2.get_Bloch_vector()
    xarr.append(x)
    yarr.append(y)
    zarr.append(z)
print(q2.get_Bloch_vector())
q2._set_Bloch_vector(Y_gate(q2.get_Bloch_vector()))
for t in range(number_time_steps):
    q2._set_Bloch_vector(amplitude_damping(q2.get_Bloch_vector(), 0.005, t))
    q2._set_Bloch_vector(phase_damping(q2.get_Bloch_vector(), 0.008, t))
    x, y, z = q2.get_Bloch_vector()
    xarr.append(x)
    yarr.append(y)
    zarr.append(z)
print(q2.get_Bloch_vector())
```
